```python
#PIECHART
loan_status = df["loan_status"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    loan_status,
    labels=loan_status.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Loan Approval Rate")
plt.show()
```

```python
edu = pd.crosstab(df["education"], df["loan_status"])

edu.plot(kind="bar", stacked=True, figsize=(7,5))

plt.title("Education vs Loan Status")
plt.xlabel("Education")
plt.ylabel("Applicants")
plt.show()
```

```python
df.head(5)
```

```python
high_loan = df[df["loan_amount"] > 24200000]["cibil_score"]
low_loan = df[df["loan_amount"] <= 24200000]["cibil_score"]

plt.figure(figsize=(7,5))

plt.boxplot([high_loan, low_loan], label=["High Loan", "Low Loan"])

plt.title("CIBIL Score by Loan Amount")
plt.xlabel("Loan Category")
plt.ylabel("CIBIL Score")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["income_annum"],
    df["loan_amount"],
    alpha=0.5
)

plt.title("Income vs Loan Amount")

plt.xlabel("Annual Income")

plt.ylabel("Loan Amount")

plt.show()
```

```python
emp = pd.crosstab(df["self_employed"],df["loan_status"])

emp.plot(kind="bar",stacked=True,figsize=(6,5))

plt.title("Self Employed vs Loan Status")

plt.ylabel("Applicants")

plt.show()
```

```python
avg_loan = df.groupby("education")["loan_amount"].mean()

avg_loan.plot(kind="bar",figsize=(6,5))

plt.title("Average Loan Amount by Education")

plt.ylabel("Average Loan Amount")

plt.show()
```

```python
import matplotlib.pyplot as plt

loan_status = df["loan_status"].value_counts()

explode = (0.08, 0)  

plt.figure(figsize=(5,5))

plt.pie(
    loan_status,
    labels=loan_status.index,
    autopct='%1.1f%%',
    startangle=90,
    explode=explode,
    shadow=True,
    wedgeprops={'edgecolor':'black','linewidth':1.2}
)

plt.title("Loan Approval Rate")
plt.axis('equal')

plt.show()
```
